Here you find a collection of conceptually equal plane waves with varying propagation and polarization directions, simulated on various numbers of cores.  
This is to show that the various ordering of patches in the lattice all can be analyzed in the same manner with the provided functions from he [get_field_data](../get_field_data.py) module.
There is otherwise not much new insight to be gained from this notebook on plane waves.   
The approaches can directly be compared to the analyses in 1D and 2D.
The settings of the waves w.r.t. wavelength and amplitude are the same.
Hence, the 2D and 1D plots look the same there as here.
The wave has a wavelength of $10\mu m$ and in all but the last scenario, the spacing of the lattice points is $1\mu m$.
Step 100 means a propagation time of $35\mu m$ and step 50 half of that.

# Propagation in x direction with polarization in y direction

```python
# Data path
scratch_dir = "/project/th-scratch/a/And.Lindner"
# Data folder
folder = "3d_plane_ypol_xdir"
# Combine
path = os.path.join(scratch_dir,folder)
path
```

```python
# Ey
component=1
```

```python
step=50
n_prc=list(range(0,1))
n_patches=[1,1,1]
patch_shape=[80,80,80]
```

```python
ey50=gfd.get_field3D(path,component,n_prc,patch_shape,n_patches,step)
```

```python
step=100
n_prc=list(range(0,1))
n_patches=[1,1,1]
patch_shape=[80,80,80]
```

```python
ey100=gfd.get_field3D(path,component,n_prc,patch_shape,n_patches,step)
```

```python
ey50.shape
```

```python
plt.figure(figsize=(20,4))
plt.subplot(1,4,1)
plt.imshow(ey50[:,40,:], cmap="seismic")
plt.subplot(1,4,2)
plt.imshow(ey100[:,40,:], cmap="seismic")
plt.subplot(1,4,3)
plt.plot(ey50[:,79,1])
plt.subplot(1,4,4)
plt.plot(ey100[:,1,1])
plt.show()
```

This is correctly propagating in the positive x-direction.
